```python
import os

def files(path):  
    for file in os.listdir(path):
        if os.path.isfile(os.path.join(path, file)):
            yield file

for file in files("."):  
    print (file)
```

```python
i=0
from PIL import Image

for fname in files("."):
    if '.png' in fname:
            try:
                i +=1
                im = Image.open(fname)
                rgb_im = im.convert('RGB')
                rgb_im.save('rayban'+str(i)+'_changed.jpg')
            except:
                print("Error "+fname)
```
